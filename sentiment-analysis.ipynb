```python
rating_zero = data['reviewText'][data['sentiment'] == 0]
rating_two= data['reviewText'][data['sentiment'] == 2]

rating_two=rating_two.astype(str)
```

```python
def list_tokenizer(sentiment): 
    sentiments = " ".join(sentiment)
    sentiments = nltk.word_tokenize(sentiments)
    return str(sentiments)
```

```python
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image  
```

```python
def word_cloud(sentiment, number): 
    wc = WordCloud(background_color = 'black', max_font_size = 50, max_words = 100)  
    wc.generate(sentiment)  
    plt.figure(figsize=(10,8))
    plt.imshow(wc, interpolation = 'bilinear')  
    plt.title(f'WordCloud for {number}')
    plt.axis('off');
```

```python
word_cloud(list_tokenizer(rating_zero), 0)
```

```python
word_cloud(list_tokenizer(rating_two), 2)
```
